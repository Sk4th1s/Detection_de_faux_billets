```python
#Import des library 
import pandas as pd
import joblib
```

```python
# Import du modèle
model = joblib.load("model_billet.pkl")
```

```python
# Import du nouveau fichier de données
new_data = pd.read_csv('billets_test.csv', sep=',')
new_data
```

```python
new_data[['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up', 'length']]
```

```python
# Applications de notre modèle sur les nouvelles données
predictions = model.predict(new_data[['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up', 'length']])
```

```python
# Résultat final
new_data["prediction"] = predictions
new_data["prediction_label"] = new_data["prediction"].map({0: "Faux billet", 1: "Vrai billet"})
print(new_data)
```

```python
# Supposons que ta colonne de prédiction s'appelle 'prediction'
counts = new_data['prediction'].value_counts()
print(counts)
```
